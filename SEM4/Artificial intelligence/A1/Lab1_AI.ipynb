```python
import numpy as np
import matplotlib.pyplot as plt
import math
from random import random

theta = np.linspace(0, 2*np.pi, 100)
x1 = np.cos(theta) 
y1 = np.sin(theta) 

plt.plot(x1, y1)

x = np.array([0,120,240,0])
y = np.array([0,120,240,0])
x = np.cos( x * np.pi / 180 )
y = np.sin( y * np.pi / 180 )

plt.plot( x, y, color='grey' )


theta = np.linspace(0, 2*np.pi, 100)
x1 = np.cos(theta)/2
y1 = np.sin(theta)/2
plt.plot(x1, y1, color='black')

valid_option = 0
for i in range(100):
    alpha = random() * (2 * np.pi)
    rand_r = math.sqrt(random())
    x, y = rand_r * np.cos(alpha), rand_r * np.sin(alpha)
    M = (x, y)
    if math.sqrt((M[0] - 0)**2 + (M[1] - 0)**2) <= 1 / 2:
        plt.plot(x, y, color='red', marker='.')
        valid_option += 1
    else:
        plt.plot(x, y, color='blue', marker='.')

print("There are " + str(valid_option) + " valid lines, probability is " + str(valid_option/100))

plt.show()
```
